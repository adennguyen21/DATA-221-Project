```python
import pandas as pd

# Load data and determine categorical features to encode
heart_disease_df = pd.read_csv("heart.csv")
print(heart_disease_df.head)
print(heart_disease_df.info())
print(heart_disease_df.describe())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Define categorical features to encode
categorical_features = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]

# Create encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown="infrequent_if_exist")

# Use ColumnTransformer to encode only categroical features
column_transformer = ColumnTransformer(transformers=[
    ("cat", encoder, categorical_features)
], remainder="passthrough")

# Transform the DataFrame
df_transformed = column_transformer.fit_transform(heart_disease_df)

# Reconstruct a readable DataFrame
encoded_feature_names = column_transformer.get_feature_names_out()
heart_disease_encoded = pd.DataFrame(df_transformed, columns=encoded_feature_names)

print(heart_disease_encoded.head())
```
